## Basic linear regression approaches

```python
df_subset_train = housing.clean_data_subset(df_train)
df_subset_test = housing.clean_data_subset(df_test)
df_subset_test.head()
```

```python
# Check to see if any fields have NaN values
housing.check_nan(df_subset_train)
housing.check_nan(df_subset_test)
```

The last thing we probably want to do is z-score each value.  This is useful to ensure all features are on similar scales.  Note that the mean and variance of each feature will need to be preserved for the test set.  Note that this should also be done for the sale price.

```python
# Normalize training features
feat_mean, feat_std = housing.calc_norm_params(df_subset_train)
df_subset_train = housing.apply_norm_params(df_subset_train, feat_mean, feat_std)
df_subset_test = housing.apply_norm_params(df_subset_test, feat_mean, feat_std)
df_subset_test.head()
```

```python
# Perform linear regression
import sklearn as skl
from sklearn import linear_model, metrics

# Get training sale prices and z-score
y_train = df_train['SalePrice']
y_mean = np.mean(y_train)
y_train = y_train - y_mean
y_std = np.std(y_train)
y_train = y_train / y_std

regr_mdl_linear = linear_model.LinearRegression()
regr_mdl_linear.fit(df_subset_train, y_train)
y_predict_train = regr_mdl_linear.predict(df_subset_train)
y_predict_test = regr_mdl_linear.predict(df_subset_test)

# Undo z-score
y_predict_train = y_predict_train * y_std + y_mean
y_predict_test = y_predict_test * y_std + y_mean

# For linear regression, the predicted value can be less than 0.
# This can cause issues for certain error metrics, such as RMSE,
# which is the case for this exercise.  To correct for this, set
# the value of all negative prices to be equal to smallest value
# in the training set.
min_val = np.min(df_train['SalePrice'])
y_predict_train[y_predict_train <= 0] = min_val
y_predict_test[y_predict_test <= 0] = min_val

# Plot linear regression results
housing.plot_fit(df_train['SalePrice'], y_predict_train, 'Linear regression -- subset')
mse = np.sqrt(metrics.mean_squared_error(np.log(df_train['SalePrice']), np.log(y_predict_train)))
print('MSE (training data): {}'.format(mse))
housing.save_submission(df_test['Id'].copy(), y_predict_test, 'linear')
```

This doesn't do great -- the RMSE for the test dataset is 0.25327.  Note that the model is over-predicting the sale price for low-price homes and over-predicting prices for high-price homes.  This could be the result of the outliers biasing the fit of the model, or the underlying relationship being non-linear.  To check for non-linearity, plot the distribution of the actual sale prices.

```python
plt.hist(df_train['SalePrice'], bins=20)
plt.xlabel('Sale price')
plt.ylabel('Counts')
plt.show()
```

As expected, the distribution of sale prices is skewed.  Applying a log transform can fix this:

```python
plt.hist(np.log(df_train['SalePrice']), bins=20)
plt.xlabel('log(Sale price)')
plt.ylabel('Counts')
plt.show()
```

```python
# Linear regression -- log-transform

# Note that the log transform has to be performed before the data is z-scored
key = 'SalePrice'
y_train = np.log(df_train[key])
y_mean = np.mean(y_train)
y_train = y_train - y_mean
y_std = np.std(y_train)
y_train = y_train / y_std

regr_mdl_linear = linear_model.LinearRegression()
regr_mdl_linear.fit(df_subset_train, y_train)
y_train_predict = regr_mdl_linear.predict(df_subset_train)
y_test_predict = regr_mdl_linear.predict(df_subset_test)

# Undo z-score and log transform
y_train_predict = np.exp(y_train_predict * y_std + y_mean)
y_test_predict = np.exp(y_test_predict * y_std + y_mean)

# Plot linear regression results
housing.plot_fit(df_train['SalePrice'], y_train_predict, 'Linear regression -- subset (log-transform)')
mse = np.sqrt(metrics.mean_squared_error(np.log(df_train['SalePrice']), np.log(y_train_predict)))
print('MSE (training data): {}'.format(mse))
housing.save_submission(df_test['Id'].copy(), y_test_predict, 'linear_logtrans')
```

This does better, giving a RMSE of 0.15112 on the test data.

However, neither of these two options do terribly well in comparison to the rest of the Kaggle submissions (somewhere around the 50th percentile in the best case).  Clearly, we need to do better.  There are a number of ways we could try to improve performance of the model:
* __More features.__  Currently we are only using a small subset of the available features (~10/80).  Including more features would probably improve performance.  However, several things would need to be done in order to make this work: (1) a method of encoding categorical variables needs to be introduced (one-hot encoding should suffice), and (2) some form of regularization will probably be important in order to prevent overfitting.
* __Use a non-linear model.__  It is pretty clear from the plots of individual features that many of these have non-linear relationships with the sale price.  Using a model that can capture these non-linearities should help.  Two possibilities include (1) a neural network, and (2) Gaussian Process regression.

## Simple feed-forward neural networks

Neural networks can be advantageous in this case, as they can model nonlinear relationships.  Try out some simple networks on the subset of data.

```python
# Simple ffnn model on subset of features
import tensorflow as tf
from tensorflow import keras

print('Tensorflow version: {}'.format(tf.__version__))

# Build model.  To start, use a smallish model
n_units = [32, 32]
def build_model(n_units, input_shape):
    model = keras.Sequential([
        tf.keras.layers.Dense(n_units[0], activation=tf.keras.activations.relu, input_shape=(input_shape,)),
        tf.keras.layers.Dense(n_units[1], activation=tf.keras.activations.relu),
        tf.keras.layers.Dense(1, activation=tf.keras.activations.linear)
    ])
    
    optimizer = keras.optimizers.Adam()
    model.compile(
        loss=tf.losses.MeanSquaredLogarithmicError(),
        optimizer=optimizer,
        metrics=[tf.keras.metrics.MeanSquaredLogarithmicError()]
    )
    
    return model
    
```

```python
# Compile model
model = build_model(n_units, len(df_subset_train.keys()))
model.summary()

# Test the model
ex_batch = df_subset_train.iloc[0:10, :].values
ex_predict = model.predict(ex_batch)
ex_predict
```

```python
# Define callback function to print a dot every 100 epochs
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('{}'.format(epoch))  # Every 100 lines, the epoch
    if epoch % 1000 == 0: print('\n')  # Every 1000 epochs, print a new line
    # print('.', end='')
```

```python
# Train the model
epochs = 2000
history = model.fit(
  df_subset_train.values, df_train['SalePrice'].values,
  epochs=epochs, validation_split=0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
# Plot history
housing.plot_history(history)
```

```python
# Predict all training data
y_train_predict = model.predict(df_subset_train.values)

# Plot results
housing.plot_fit(df_train['SalePrice'], y_train_predict, 'FFNN -- subset')
mse = np.sqrt(metrics.mean_squared_error(np.log(df_train['SalePrice']), np.log(y_train_predict)))
print('MSE (training data): {}'.format(mse))
```

```python
# Predict test data and save
y_test_predict = model.predict(df_subset_test.values)
housing.save_submission(df_test['Id'].copy(), y_test_predict, 'ffnn_32_2000')
```

This does slightly better, resulting in a test set error of ~0.146.  Better than linear regression, but not by much.  Try increasing the size of the network.

```python
# Compile model
n_units = [128, 128]
model = build_model(n_units, len(df_subset_train.keys()))

# Train the model
epochs = 2000
history = model.fit(
  df_subset_train.values, df_train['SalePrice'].values,
  epochs=epochs, validation_split=0.02, verbose=0,
  callbacks=[PrintDot()])
```

```python
# Plot history
housing.plot_history(history)

# Predict all training data
y_train_predict = model.predict(df_subset_train.values)

# Plot results
housing.plot_fit(df_train['SalePrice'], y_train_predict, 'FFNN -- subset')
mse = np.sqrt(metrics.mean_squared_error(np.log(df_train['SalePrice']), np.log(y_train_predict)))
print('MSE (training data): {}'.format(mse))

# Predict test data and save
y_test_predict = model.predict(df_subset_test.values)
housing.save_submission(df_test['Id'].copy(), y_test_predict, 'ffnn_128_2000')
```

This is starting to do better on the training data, but has slightly higher error on the training set (~0.150), which we might expect from the number of parameters.  Let's see if we can over-fit even more:

```python
# Define function to bulid the model.  Allow an arbitrary number of layers to be specified.
def build_model(n_units, input_shape):
    # Add the first layer with the appropriate input shape
    layers = [tf.keras.layers.Dense(
        n_units[0],
        activation=tf.keras.activations.relu,
        input_shape=(input_shape,)
    )]
    # Iterate over the remaining layers
    for u in n_units[1:]:
        layers.append(
            tf.keras.layers.Dense(u, activation=tf.keras.activations.relu)
        )
    # Add the output layer
    layers.append(
        tf.keras.layers.Dense(1, activation=tf.keras.activations.linear)
    )
    
    model = keras.Sequential(layers)
    optimizer = keras.optimizers.Adam()
    model.compile(
        loss=tf.losses.MeanSquaredLogarithmicError(),
        optimizer=optimizer,
        metrics=[tf.keras.metrics.MeanSquaredLogarithmicError()]
    )
    
    return model
```

```python
n_units = [128, 128, 128, 128]
model = build_model(n_units, len(df_subset_train.keys()))
model.summary()

# Train the model
epochs = 10000
history = model.fit(
  df_subset_train.values, df_train['SalePrice'].values,
  epochs=epochs, validation_split=0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
# Plot history
housing.plot_history(history)

# Predict all training data
y_train_predict = model.predict(df_subset_train.values)

# Plot results
housing.plot_fit(df_train['SalePrice'], y_train_predict, 'FFNN -- subset (4 layers, 128 units/layer)')
mse = np.sqrt(metrics.mean_squared_error(np.log(df_train['SalePrice']), np.log(y_train_predict)))
print('MSE (training data): {}'.format(mse))

# Predict test data and save
y_test_predict = model.predict(df_subset_test.values)
housing.save_submission(df_test['Id'].copy(), y_test_predict, 'ffnn_4_128_10000')
```

This is clearly overfitting -- the accuracy on the training set is relatively low (~0.1), but the error of the model on the test set is higher (~0.2).  Also note that the validation error is starting to creep up fairly quickly -- over the first 2000 epochs -- which is another indication we're overfitting.

At this point, we have seen small gains from using the FFNN.  However, we are currently using only a small subset of the features, and only those which are somewhat continuously-valued.  In order to make further gains, we should probably include additional features.  This will require dealing with categorical variables.
